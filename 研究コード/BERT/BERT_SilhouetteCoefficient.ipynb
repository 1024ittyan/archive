```python
import pandas as pd

df_vec = pd.read_csv('/content/drive/MyDrive/BERT/BERT_vecs.csv')
```

```python
df_vec
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import csv

silhouette_scores = []
for k in range(2, 30):
    print(k)
    kmeans = KMeans(n_clusters=k,max_iter=1000,n_init=10)
    kmeans.fit(df_vec)
    silhouette_scores.append(silhouette_score(df_vec, kmeans.labels_))
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(range(2, 30), silhouette_scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette coefficient')
plt.xticks(range(2, 30, 1))
plt.title(f"BERT_SilhouetteCoefficient")
plt.show()
best_k = np.argmax(silhouette_scores) + 2
print(best_k)
```
